**2.Создаем массвы с помощью обычных списков и с использользованием функий NumPy**

```python
massive1 = np.array([1,2,3,4])
print(massive1) #одномерный массив
```

```python
massive2 = np.array([[1,2,3],[3,4,6]])
print(massive2) #двумернный массив
massive2floati = np.array([[1.2,2.3,3.9],[3.123213,4.1232138103812,6.23193821031]])
print(massive2floati) #двумернный массив с флоутами
```

*Используем функции NumPy*

```python
massive3 = np.arange(1,10,2)
print(massive3)
```

```python
massive4 = np.arange(10)
print(massive4)
```

```python
massive5 = np.zeros((2,3))
print(massive5)
```

```python
massive6 = np.ones((2,3))
print(massive6)
```

```python
massive7 = np.eye(3)
print(massive7)
```

**4.Примеры индексирования и срезов**

```python
element = massive2floati[1,2]
print("Отдельный эллемент:", element) #получаем отдельный элемент
```

```python
podmasive = massive2floati[:1,2:]
print("Подмасив:", podmasive) #получаем подмасив срезом
```

```python
row = massive2floati[:1, :]
print("Строка:", row) #выводим строку
```

```python
column = massive2floati[: ,1:]
print("Столбик:", column) #выводим столбец
```
